```python
import glob

all_text = []
for file_path in glob.glob('./texts/*.txt'):
    file = open(file_path, 'r', encoding='utf8')
    all_text.append(file.read())

all_text = '\n\n\n'.join(all_text)
print(len(all_text))
```

```python
%cd myLLMs
```

```python
from BPE import BPE

tokenizer = BPE(2000)
tokenizer.fit(all_text)
```

```python
print(tokenizer.id2token)
```

```python
tokenizer.save('data/tokenizer.dill')
```
